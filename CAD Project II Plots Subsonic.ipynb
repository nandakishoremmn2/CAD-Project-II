```python
%pylab inline
from scipy.io import *
```

### Loading all data

```python
datafile = 'subsonic.mat'
```

```python
data = loadmat(datafile)
x = data["x"]
y = data["y"]
p = data["p"]
v = data["v"]
u = data["u"]
d = data["d"]
residue = data["residue"]
iterations = data["iterations"]
```

## Convergence History

```python
plot(np.abs(residue/residue[0]))
yscale("log")
title("Convergence History")
print np.min(residue/residue[0]), residue[-1]/residue[0]
print "Total iterations = %d"%iterations
```

## Mach Contours

```python
M = sqrt( (u**2+v**2)/(1.4*p/d) )
CS = contourf(x, y, M, 100)
colorbar(CS)
title("Mach contours")
axis("image")
```

## Pressure Contours

```python
CS = contourf(x, y, p, 100)
colorbar(CS)
title("Pressure contours")
axis("image")
np.min(p)
```

## Density Contours

```python
CS = contourf(x, y, d, 100)
colorbar(CS)
title("Density contours")
axis("image")
```

## Velocity Contours

```python
CS = contourf(x, y, u, 100)
colorbar(CS)
title("$u$ velocity contours")
axis("image")
print np.max(u), np.min(u)
```

```python
CS = contourf(x, y, v, 100)
colorbar(CS)
title("$v$ velocity contours")
axis("image")
print np.min(v), np.max(v)
```

```python
CS = contourf(x, y, sqrt(u**2+v**2), 100)
colorbar(CS)
title("resultant velocity ($\sqrt{u^2+v^2}$) contours")
axis("image")
```

```python
alt = lambda x: x[0:24:4,24:74:2]
quiver(*map(alt, [x, y, u, v]))
title("Velocity vectors in the proximity of the bump")
axis("image")
```
